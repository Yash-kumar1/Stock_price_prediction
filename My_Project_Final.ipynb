```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import math
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
url='/content/TSLA.csv'
dataset =pd.read_csv(url, encoding='unicode_escape')
dataset.head()
```

```python
dataset.tail()
```

```python
x = len(dataset['Date'])
```

```python
x = dataset[['High','Low','Open','Volume']].values
y = dataset['Close'].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

# Linear Regression

```python
Regressor = LinearRegression()
Regressor.fit(x_train,y_train)
```

```python
print(Regressor.coef_)
print('----------------------------')

print(Regressor.intercept_)
```

```python
predicted = Regressor.predict(x_test)
```

```python
print(predicted)
```

```python
dframe = pd.DataFrame({'Actual':y_test.flatten(),'Predicted':predicted.flatten()})

dframe.head(25)
```

```python
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_test,predicted))
print('Mean Squared Error: ',metrics.mean_squared_error(y_test, predicted))
print('Root Mean Squared Error: ',math.sqrt(metrics.mean_squared_error(y_test, predicted)))
```

```python
graph = dframe.head(30)
# plt.xlabel("Average Pulse")
graph.plot(kind='bar', xlabel='Days', ylabel='Prices')
```

# SVR

```python
SVR_Regressor = SVR()
SVR_Regressor.fit(x_train,y_train)
```

```python
SVR_predicted = SVR_Regressor.predict(x_test)
```

```python
print(SVR_predicted)
```

```python
SVR_dframe = pd.DataFrame({'SVR Actual':y_test.flatten(),'SVR Predicted':SVR_predicted.flatten()})

SVR_dframe.head(25)
```

```python
print('SVR Mean Absolute Error: ',metrics.mean_absolute_error(y_test,SVR_predicted))
print('SVR Mean Squared Error: ',metrics.mean_squared_error(y_test, SVR_predicted))
print('SVR Root Mean Squared Error: ',math.sqrt(metrics.mean_squared_error(y_test, SVR_predicted)))
```

```python
SVR_graph = SVR_dframe.head(30)
SVR_graph.plot(kind='bar', xlabel='Days', ylabel='Prices')
```

# Decision Tree

```python
RForest_Regressor = RandomForestRegressor()
```

```python
RForest_Regressor.fit(x_train,y_train)
```

```python
RForest_predicted = RForest_Regressor.predict(x_test)
```

```python
RForest_predicted = RForest_Regressor.predict(x_test)
```

```python
RForest_dframe = pd.DataFrame({'Random Forest Regressor Actual':y_test.flatten(),'Random Forest Predicted':RForest_predicted.flatten()})

RForest_dframe.head(25)
```

```python
print('RForest Mean Absolute Error: ',metrics.mean_absolute_error(y_test,RForest_predicted))
print('RForest Mean Squared Error: ',metrics.mean_squared_error(y_test, RForest_predicted))
print('RForest Root Mean Squared Error: ',math.sqrt(metrics.mean_squared_error(y_test, RForest_predicted)))
```

```python
RForest_graph = RForest_dframe.head(30)
RForest_graph.plot(kind='bar', xlabel='Days', ylabel='Prices')
```
